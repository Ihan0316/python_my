```python
def display_menu():
    print("\n🔹 리스트 관리 시스템 🔹")
    print("1. 추가")
    print("2. 수정")
    print("3. 삭제")
    print("4. 조회")
    print("5. 검색")
    print("6. 종료")

def add_item(my_list):
    item = input("추가할 항목을 입력하세요: ")
    my_list.append(item)
    print(f"'{item}' 항목이 추가되었습니다.")

def modify_item(my_list):
    index = int(input("수정할 항목의 인덱스를 입력하세요: "))
    if 0 <= index < len(my_list):
        new_item = input("새 항목을 입력하세요: ")
        my_list[index] = new_item
        print(f"인덱스 {index}의 항목이 '{new_item}'(으)로 수정되었습니다.")
    else:
        print("잘못된 인덱스입니다.")

def delete_item(my_list):
    index = int(input("삭제할 항목의 인덱스를 입력하세요: "))
    if 0 <= index < len(my_list):
        removed_item = my_list.pop(index)
        print(f"'{removed_item}' 항목이 삭제되었습니다.")
    else:
        print("잘못된 인덱스입니다.")

def view_items(my_list):
    if my_list:
        print("리스트 항목:")
        for i, item in enumerate(my_list):
            print(f"{i}: {item}")
    else:
        print("리스트가 비어 있습니다.")

def search_item(my_list):
    search_term = input("검색할 항목을 입력하세요: ")
    results = [item for item in my_list if search_term in item]
    if results:
        print("검색 결과:")
        for item in results:
            print(item)
    else:
        print("검색 결과가 없습니다.")

def main():
    my_list = []
    while True:
        display_menu()
        choice = input("메뉴 선택(숫자로 해주세요): ")
        if choice == '1':
            add_item(my_list)
        elif choice == '2':
            modify_item(my_list)
        elif choice == '3':
            delete_item(my_list)
        elif choice == '4':
            view_items(my_list)
        elif choice == '5':
            search_item(my_list)
        elif choice == '6':
            print("프로그램을 종료합니다.")
            break
        else:
            print("잘못된 선택입니다. 다시 시도하세요.")

if __name__ == "__main__":
    main()
```

```python
def display_menu():
    print("\n📞 전화번호부 관리 시스템 📞")
    print("1. 추가")
    print("2. 수정")
    print("3. 삭제")
    print("4. 조회")
    print("5. 검색")
    print("6. 종료")

def add_phonenumber(phonebook):
    name = input("이름을 입력하세요: ")
    phonenumber = input("전화번호를 입력하세요: ")
    phonebook[name] = phonenumber
    print(f"'{name}'의 전화번호가 추가되었습니다.")

def modify_phonenumber(phonebook):
    name = input("수정할 이름을 입력하세요: ")
    if name in phonebook:
        new_phonenumber = input("새 전화번호를 입력하세요: ")
        phonebook[name] = new_phonenumber
        print(f"'{name}'의 전화번호가 '{new_phonenumber}'(으)로 수정되었습니다.")
    else:
        print("해당 이름이 전화번호부에 없습니다. 다시 검색 부탁드립니다.")

def delete_phonenumber(phonebook):
    name = input("삭제할 이름을 입력하세요: ")
    if name in phonebook:
        del phonebook[name]
        print(f"'{name}'의 전화번호가 삭제되었습니다.")
    else:
        print("해당 이름이 전화번호부에 없습니다.")

def view_phonenumbers(phonebook):
    if phonebook:
        print("전화번호부:")
        for name, phonenumber in phonebook.items():
            print(f"{name}: {phonenumber}")
    else:
        print("전화번호부가 비어 있습니다.")

def search_phonenumber(phonebook):
    search_term = input("검색할 이름을 입력하세요: ")
    results = {name: phonenumber for name, phonenumber in phonebook.items() if search_term in name}
    if results:
        print("검색 결과:")
        for name, phonenumber in results.items():
            print(f"{name}: {phonenumber}")
    else:
        print("검색 결과가 없습니다.")

def main():
    phonebook = {}
    while True:
        display_menu()
        choice = input("메뉴 선택(숫자로 해주세요): ")
        if choice == '1':
            add_phonenumber(phonebook)
        elif choice == '2':
            modify_phonenumber(phonebook)
        elif choice == '3':
            delete_phonenumber(phonebook)
        elif choice == '4':
            view_phonenumbers(phonebook)
        elif choice == '5':
            search_phonenumber(phonebook)
        elif choice == '6':
            print("프로그램을 종료합니다.")
            break
        else:
            print("잘못된 선택입니다. 다시 시도하세요.")

if __name__ == "__main__":
    main()
```
